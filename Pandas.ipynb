# Series Creation with Dictionary

```python
h = pd.Series({'a':1, 'b':2,'c':3}, name = "Dict")
h
```

```python
#If the series is created via a dictionary
#then if you change/add the index then it will show the index, however the values will be filled with null values
#Null values in pandas are known as NaN(Not a Number)

J = pd.Series({"a":1,"b":2,"c":3},index=[2.3,4.3,5.3],name="Dict")
J
```
